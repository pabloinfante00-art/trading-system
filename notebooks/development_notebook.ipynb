## 3. Data Transformation (ETL — Transform)

We develop the pipeline step-by-step for **AAPL only** first.
Once we verify it works, section 3.7 runs it for all 5 tickers automatically.

```python
# Filter for AAPL — used for exploration in sections 3.1 to 3.5
ticker = 'AAPL'
df = prices_df[prices_df['Ticker'] == ticker].copy()
print(f'Filtered {len(df):,} rows for {ticker}')

company_info = companies_df[companies_df['Ticker'] == ticker]
print(f'Company: {company_info.iloc[0]["Company Name"]}')
```

### 3.1 Cleaning

```python
# Parse dates and sort
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())

# Drop rows with missing Close price — unusable without a price
df = df.dropna(subset=['Close'])

# Fill missing Volume with 0 (SimFin free tier omits Volume for some dates)
if 'Volume' in df.columns:
    df['Volume'] = df['Volume'].fillna(0)

# Use Close as a fallback when High/Low are missing
if 'High' in df.columns:
    df['High'] = df['High'].fillna(df['Close'])
if 'Low' in df.columns:
    df['Low'] = df['Low'].fillna(df['Close'])

# Remove duplicate dates (keep the last entry)
df = df.drop_duplicates(subset=['Date'], keep='last').reset_index(drop=True)

print(f'\nCleaned data: {len(df):,} rows')
df.head()
```

### 3.2 Feature Engineering

We create **technical indicators** that the ML model will use as input features:
- **Returns**: Daily percentage change in price
- **SMA_5, SMA_20**: Simple Moving Averages (short/medium term trends)
- **EMA_12**: Exponential Moving Average (weights recent prices more)
- **RSI_14**: Relative Strength Index (momentum — overbought/oversold)
- **Volatility_20**: Rolling standard deviation of returns
- **Volume_Change**: Daily change in trading volume
- **High_Low_Range**: Intraday price range

```python
# Daily Returns
df['Returns'] = df['Close'].pct_change()

# Simple Moving Averages
df['SMA_5']  = df['Close'].rolling(window=5).mean()
df['SMA_20'] = df['Close'].rolling(window=20).mean()

# Exponential Moving Average
df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()

# RSI (Relative Strength Index)
delta    = df['Close'].diff()
gain     = delta.where(delta > 0, 0.0)
loss     = (-delta).where(delta < 0, 0.0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
df['RSI_14'] = 100 - (100 / (1 + rs))

# Volatility
df['Volatility_20'] = df['Returns'].rolling(window=20).std()

# Volume Change
if 'Volume' in df.columns and df['Volume'].sum() > 0:
    df['Volume_Change'] = df['Volume'].pct_change().replace([np.inf, -np.inf], 0).fillna(0)
else:
    df['Volume_Change'] = 0.0

# High-Low Range
df['High_Low_Range'] = df['High'] - df['Low']

print('Features created!')
df[['Date', 'Close', 'Returns', 'SMA_5', 'SMA_20', 'RSI_14', 'Volatility_20']].tail(10)
```

### 3.4 Create Target Variable

Our target is **binary classification**:
- `1` = tomorrow's Close price is **higher** than today's (UP)
- `0` = tomorrow's Close price is **lower or equal** to today's (DOWN)

```python
# 1 if next day's Close > today's Close, else 0
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

print('Target distribution:')
print(df['Target'].value_counts())
print(f'\nUp days:   {df["Target"].sum()} ({df["Target"].mean()*100:.1f}%)')
print(f'Down days: {(df["Target"] == 0).sum()} ({(1-df["Target"].mean())*100:.1f}%)')
```

## 4. Machine Learning Model

We use **Logistic Regression** for binary classification (UP/DOWN) on all 5 tickers.

**Why Logistic Regression?**
- Simple and interpretable
- Works well as a baseline for binary classification
- Fast to train
- Outputs probabilities (confidence)

### 4.1 Train a Model for Each Ticker

We loop over all 5 tickers, load the processed CSV from section 3.7, and train one
Logistic Regression model per company. Results are stored in the `models` dict.

```python
models = {}  # stores trained model, scaler, predictions and metrics for each ticker

for t in TICKERS:
    path = os.path.join(PROCESSED_DIR, f'{t}_processed.csv')
    df_t = pd.read_csv(path)

    X = df_t[FEATURE_COLUMNS].replace([np.inf, -np.inf], np.nan)
    y = df_t['Target']
    mask = X.notna().all(axis=1)
    X, y = X[mask], y[mask]

    # shuffle=False: train on past, test on future (time-series rule)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=False
    )

    # Fit scaler only on training data to avoid leakage into the test set
    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s  = scaler.transform(X_test)

    model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
    model.fit(X_train_s, y_train)

    y_pred = model.predict(X_test_s)

    models[t] = {
        'model':            model,
        'scaler':           scaler,
        'y_test':           y_test,
        'y_pred':           y_pred,
        'accuracy':         accuracy_score(y_test, y_pred),
        'confusion_matrix': confusion_matrix(y_test, y_pred),
        'importance':       pd.DataFrame({
            'Feature':     FEATURE_COLUMNS,
            'Coefficient': model.coef_[0]
        }).sort_values('Coefficient', key=abs, ascending=True),
    }

    print(f'{t}: accuracy = {models[t]["accuracy"]:.1%}')

print('\nAll models trained.')
```

### 4.2 Accuracy Summary

```python
summary = pd.DataFrame({
    t: {
        'Accuracy':     f"{models[t]['accuracy']:.1%}",
        'Train rows':   int(len(pd.read_csv(os.path.join(PROCESSED_DIR, f'{t}_processed.csv'))) * 0.8),
        'Test rows':    int(len(pd.read_csv(os.path.join(PROCESSED_DIR, f'{t}_processed.csv'))) * 0.2),
    }
    for t in TICKERS
}).T

print('Model accuracy across all tickers:\n')
print(summary.to_string())
print('\nNote: ~50% accuracy is expected — stock prediction is hard.')
```

### 4.3 Confusion Matrices — All Tickers

Each heatmap shows how well the model distinguishes UP from DOWN days.
Row = actual class, column = predicted class.

```python
fig, axes = plt.subplots(1, len(TICKERS), figsize=(22, 4))
fig.suptitle('Confusion Matrices — All Tickers', fontsize=14, y=1.02)

for i, t in enumerate(TICKERS):
    cm  = models[t]['confusion_matrix']
    acc = models[t]['accuracy']

    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=['DOWN', 'UP'],
        yticklabels=['DOWN', 'UP'],
        ax=axes[i],
        cbar=False,
        linewidths=0.5,
        annot_kws={'size': 14, 'weight': 'bold'},
    )
    axes[i].set_title(f'{t}\nAcc: {acc:.1%}', fontsize=11)
    axes[i].set_xlabel('Predicted', fontsize=9)
    axes[i].set_ylabel('Actual' if i == 0 else '', fontsize=9)
    axes[i].tick_params(labelsize=9)

plt.tight_layout()
plt.savefig(os.path.join(PROCESSED_DIR, 'confusion_matrices.png'), dpi=150, bbox_inches='tight')
plt.show()
```

### 4.4 Feature Importance — All Tickers

Logistic Regression coefficients indicate which features push the model toward UP (green, positive)
or DOWN (red, negative). Larger absolute value = stronger influence.

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Feature Importance (Logistic Regression Coefficients) — All Tickers', fontsize=14)
axes_flat = axes.flatten()

for i, t in enumerate(TICKERS):
    imp    = models[t]['importance'].copy()
    colors = ['#2ecc71' if c > 0 else '#e74c3c' for c in imp['Coefficient']]

    sns.barplot(
        data=imp,
        y='Feature',
        x='Coefficient',
        palette=colors,
        ax=axes_flat[i],
        orient='h',
    )
    axes_flat[i].axvline(0, color='black', linewidth=0.8, linestyle='--')
    axes_flat[i].set_title(t, fontsize=12, fontweight='bold')
    axes_flat[i].set_xlabel('Coefficient', fontsize=9)
    axes_flat[i].set_ylabel('')
    axes_flat[i].tick_params(labelsize=9)

# Hide the unused 6th subplot (2x3 grid, only 5 tickers)
axes_flat[5].set_visible(False)

plt.tight_layout()
plt.savefig(os.path.join(PROCESSED_DIR, 'feature_importance.png'), dpi=150, bbox_inches='tight')
plt.show()
```

### 4.5 Export All Models

Save the model and scaler for each ticker as `.pkl` files.
The web app needs **both** — the scaler to normalize live data, then the model to predict.

```python
os.makedirs(MODELS_DIR, exist_ok=True)

for t in TICKERS:
    model_path  = os.path.join(MODELS_DIR, f'{t}_model.pkl')
    scaler_path = os.path.join(MODELS_DIR, f'{t}_scaler.pkl')
    joblib.dump(models[t]['model'],  model_path)
    joblib.dump(models[t]['scaler'], scaler_path)
    print(f'{t}: model -> {model_path} | scaler -> {scaler_path}')

print('\nAll models exported.')
```
